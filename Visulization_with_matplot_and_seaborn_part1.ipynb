```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
yield_apple = [0.895, 0.91,0.919,0.926,0.929,0.931]
```

```python
plt.plot(yield_apple);
```

```python
year=[2013,2014,2015,2016,2017,2018]
yield_apple = [0.895, 0.91,0.919,0.926,0.929,0.931]
plt.plot(year,yield_apple);
plt.xlabel('Year');
plt.ylabel('Yield (tons per hectare)');
```

Ploting multiple lines

```python
year= range(2000,2012)
yield_apple = [0.895, 0.91,0.919,0.926,0.929,0.931,0.934,0.936,0.941,0.955,0.962,0.98]
yield_orange = [0.962,0.967,0.93,0.91,0.94,0.97,0.91,0.93,0.98,0.91,0.95,0.99]
plt.plot(year,yield_apple);
plt.plot(year,yield_orange);
plt.xlabel('Year');
plt.ylabel('Yield (tons per hectare)');
```

chart title and legend

```python
plt.plot(year,yield_apple);
plt.plot(year,yield_orange);
plt.xlabel('Year');
plt.ylabel('Yield (tons per hectare)');
plt.title('Crop Yields in Kanto');
plt.legend(['Apple','Orange']);
```

Line markers

```python
plt.plot(year,yield_apple, marker = 'o');
plt.plot(year,yield_orange , marker= 'x');
plt.xlabel('Year');
plt.ylabel('Yield (tons per hectare)');
plt.title('Crop Yields in Kanto');
plt.legend(['Apple','Orange']);
```

```python
plt.plot(year,yield_apple, marker = 'o', c='b',ls='-',lw=2, ms=8, mew=2, mec = 'navy');
plt.plot(year,yield_orange , marker= 'x', c='r',ls='--',lw=3, ms=10, mew=0.5);
plt.xlabel('Year');
plt.ylabel('Yield (tons per hectare)');
plt.title('Crop Yields in Kanto');
plt.legend(['Apple','Orange']);
```

```python
plt.plot(year,yield_apple, 's-b');
plt.plot(year,yield_orange ,'o--r');
plt.xlabel('Year');
plt.ylabel('Yield (tons per hectare)');
plt.title('Crop Yields in Kanto');
plt.legend(['Apple','Orange']);
```
