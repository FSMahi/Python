Multiple Histogram

```python
setosa_df = flower_df[flower_df.species == 'setosa']
versicolor_df = flower_df[flower_df.species == 'versicolor']
virginica_df = flower_df[flower_df.species == 'virginica']
```

```python
setosa_df
```

```python
plt.hist(setosa_df.sepal_width, bins=np.arange(2,5,.25), alpha=0.4, label='Setosa');
plt.hist(versicolor_df.sepal_width, bins=np.arange(2,5,.25), alpha=0.4, label='Versicolor');
plt.legend(['setosa','versicolor']);
```

```python
plt.hist(setosa_df.sepal_width, bins=np.arange(2,5,.25), alpha=0.5, label='Setosa');
plt.hist(versicolor_df.sepal_width, bins=np.arange(2,5,.25), alpha=0.4, label='Versicolor');
plt.hist(virginica_df.sepal_width, bins=np.arange(2,5,.25), alpha=0.5, label='Virginica');
plt.legend(['setosa','versicolor','virginica']);
```

```python
plt.title('Distribution of Sepal Width')
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width], bins=np.arange(2,5,.25),  stacked=True);
plt.legend(['setosa','versicolor','virginica']);
```

**Bar chart**

```python
years= range(2000,2006)
apples = [0.35, 0.6, 0.9, 0.926, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
```

```python
plt.plot(years,apples);
```

```python
plt.bar(years,apples);
```

```python
plt.bar(years,apples, label='Apples');
plt.bar(years,oranges, label='Oranges');
plt.legend();
```

```python
plt.bar(years,apples);
plt.plot(years,apples, 'o--r');
plt.title('Apples');
```

```python
plt.bar(years,apples, label='Apples');
plt.bar(years,oranges, label='Oranges', bottom= apples);
plt.legend();
```

# Heatmap

```python
flights_df = df_flight.pivot(index='month', columns='year', values='passengers')
flights_df
```

```python
plt.title('No. of Passengers');
sns.heatmap(flights_df);
```

```python
plt.title('No. of Passengers');
sns.heatmap(flights_df,fmt='d' ,annot=True,cmap='Blues');
```
